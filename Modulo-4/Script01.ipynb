## Formateando la salida de un objeto `datetime`
Recurso : https://strftime.org/

```python
# Creacion de un objeto datetime a partir de un objeto date y de un objeto time
fecha = datetime.date(2026,3,26)
hora = datetime.time(17,25,29)

dt = datetime.datetime.combine(fecha, hora)
dt
```

```python
# Metodo strftime de una instancia de la clase datetime.datetime
formato1 = "%B %A"
dt.strftime(formato1)
```

```python
formato2 = "%w - %H -- %M"
dt.strftime(formato2)
```
